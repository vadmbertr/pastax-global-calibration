```python
eastward_mask = (gdp_ds.eastward_stress > 0).compute()
northward_mask = (gdp_ds.northward_stress > 0).compute()

ep_np_mask = eastward_mask & northward_mask
ep_nn_mask = eastward_mask & ~northward_mask
en_nn_mask = ~eastward_mask & ~northward_mask
en_np_mask = ~eastward_mask & northward_mask

plt.scatter(
    np.abs(gdp_ds.eastward_stress[ep_np_mask]), np.abs(gdp_ds.northward_stress[ep_np_mask]), 
    s=1, alpha=0.5, label="ep_np_mask"
)
plt.scatter(
    np.abs(gdp_ds.eastward_stress[ep_nn_mask]), np.abs(gdp_ds.northward_stress[ep_nn_mask]), 
    s=1, alpha=0.5, label="ep_nn_mask"
)
plt.scatter(
    np.abs(gdp_ds.eastward_stress[en_nn_mask]), np.abs(gdp_ds.northward_stress[en_nn_mask]), 
    s=1, alpha=0.5, label="en_nn_mask"
)
plt.scatter(
    np.abs(gdp_ds.eastward_stress[en_np_mask]), np.abs(gdp_ds.northward_stress[en_np_mask]), 
    s=1, alpha=0.5, label="en_np_mask"
)

plt.xscale("log")
plt.yscale("log")
plt.legend()

plt.show()
```

# Environmental variables ratio

Velocities are represented as complex numbers: $x = u_x + i v_x$, giving the polar form $x = |x| e^{i \theta_x}$.

Two paired-velocities are compared by taking their ratio: $r = x / y$, giving in the polar form: $r = |x| / |y| e^{i(\theta_x - \theta_y)}$.

```python
def complex_ratio_plot(z1, z2, label_z1, label_z2, cmap="inferno", eps=1e-8):
    mask = (z2 > eps).compute()
    z1 = z1[mask]
    z2 = z2[mask]

    r = z1 / z2
    df = xr.Dataset({"x": r.real, "y": r.imag}).to_dask_dataframe()

    lim = np.nanquantile(np.abs(r), 0.95)

    canvas_size = int(600 * lim / 2)

    points = rasterize(
        hv.Points(df, kdims=["x", "y"]), aggregator=ds.count()
    ).opts(
        cmap=cmap,
        cnorm="log",
        xlim=(-lim, lim),
        ylim=(-lim, lim),
        width=canvas_size,
        height=canvas_size,
        xlabel="Re(y / x)",
        ylabel="Im(y / x)",
        title=f"{label_z1} / {label_z2}",
        colorbar=True,
        frame_width=400,
        frame_height=400,
    )

    theta = np.linspace(0, 2 * np.pi, 400)
    unit_circle = hv.Curve(
        (np.cos(theta), np.sin(theta))
    ).opts(color="black", line_dash="dashed", line_width=1)

    x_axis = hv.HLine(0).opts(color="black", line_dash="dashed", line_width=1)
    y_axis = hv.VLine(0).opts(color="black", line_dash="dashed", line_width=1)

    perfect = hv.Points([(1, 0)]).opts(color="cyan", size=8, marker="x")

    return points * unit_circle * x_axis * y_axis * perfect
```

```python
drifter = gdp_ds.ve + 1j * gdp_ds.vn
geos = gdp_ds.ugos + 1j * gdp_ds.vgos
wind = gdp_ds.eastward_wind + 1j * gdp_ds.northward_wind
stress = gdp_ds.eastward_stress + 1j * gdp_ds.northward_stress
stokes = gdp_ds.VSDX + 1j * gdp_ds.VSDY
```

```python
complex_ratio_plot(drifter, geos, "Drifter velocity", "Geostrophic current")
```

```python
complex_ratio_plot(drifter, stokes, "Drifter velocity", "Stokes drift")
```

```python
complex_ratio_plot(drifter, wind, "Drifter velocity", "10 m wind")
```

```python
complex_ratio_plot(drifter, stress, "Drifter velocity", "Wind stress")
```

```python
complex_ratio_plot(stress, wind, "Wind stress", "10 m wind")
```

```python
complex_ratio_plot(stokes, wind, "Stokes drift", "10 m wind")
```

This notebook applies some dummy cleaning to the interpolated data, for now only removing incoherent wind stress values.

```python
import numpy as np
import xarray as xr
```

```python
from_datetime_str = "1994-06-01"
to_datetime_str = "2025-08-01"
```

```python
gdp_ds = xr.open_zarr(f"data/gdp_interp_{from_datetime_str}_{to_datetime_str}.zarr.old")
```

```python
len(gdp_ds.points)
```

```python
gdp_ds = gdp_ds.where(
    (((gdp_ds.eastward_stress ** 2 + gdp_ds.northward_stress ** 2) ** 0.5) < 10).compute(), drop=True
)
```

```python
len(gdp_ds.points)
```

```python
gdp_ds["day_of_year"] = gdp_ds.time.dt.dayofyear.astype(np.uint16)
```

```python
gdp_ds["hour_of_day"] = gdp_ds.time.dt.hour.astype(np.uint8)
```

```python
gdp_ds.to_zarr(f"data/gdp_interp_clean_{from_datetime_str}_{to_datetime_str}.zarr", mode="w")
```
